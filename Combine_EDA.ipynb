```python
import glob
import pandas as pd

# define the path, and grab all csv files
csv_files = glob.glob("/content/DAT490_Capstone_Combine/Data/Combine/*.csv")

# create empty list to store data frames
all_df = []

# loop through all dfs and store in list
for file in csv_files:
  df = pd.read_csv(file)
  all_df.append(df)

# merge list into one df
combine_df = pd.concat(all_df, ignore_index = True)
```
